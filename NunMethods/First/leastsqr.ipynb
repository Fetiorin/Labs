````python
def leastSqr(x, y, n):
    H = np.zeros((len(x), n+1))
    for i in range(len(x)):
        for j in range(n+1):
            H[i][j] = x[i] ** j
    HH = np.dot(H.T, H)
    Hy = np.dot(H.T, y)
    a = np.linalg.solve(HH, Hy)
    return lambda x: np.dot(a, [x ** i for i in range(len(a))])
````

````python
def f(x):
    return np.sin(x)
````

````python
x = np.linspace(-5, 5, num=3000)
y = f(x) + np.random.normal(0, 0.25, 3000)
a = leastSqr(x, y, 6)

plt.subplots(figsize=(10, 6))
plt.plot(x, y, 'g.')

xx = np.linspace(-5.5, 5.5, num=100)
plt.plot(xx, a(xx), 'b-')
#plt.plot(xx, f(xx), 'b--')
plt.show()
````
